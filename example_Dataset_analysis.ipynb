### From here onwards, we start to test xgBoost on a simple dataset with NaN values

```python
gse = GEOparse.get_GEO(filepath="D:\My BUET Documents\October 18\BioInformatics Algorithms\project\Datasets\FINAL_dataset\gds-soft\GDS12.soft.gz")
GDS12 = gse.table
GDS12
```

```python
zeros = [0,0,0,0]
ones = [1,1,1,1]
Label = zeros + ones
Label
```

```python
GDS12 = GDS12.T
GDS12 = GDS12.iloc[2:,0:]
GDS12['Label'] = Label
GDS12
```

```python
from sklearn.utils import shuffle
GDS12 = shuffle(GDS12)
GDS12
```

```python
GDS12_wNaN = GDS12
GDS12_woNaN = GDS12.fillna(0)
```

```python
GDS12_wNaN.to_csv("GDS12wNan.csv",index=False)
GDS12_woNaN.to_csv("GDS12woNan.csv",index=False)
```

```python
Train_data, Test_data = GDS12_wNaN[:6], GDS12_wNaN[6:]
X_data_tr, y_tr = Train_data.drop('Label', axis=1), Train_data['Label']
X_data_te, y_te = Test_data.drop('Label', axis=1), Test_data['Label']
```

```python
Train_data.shape, Test_data.shape, X_data_tr.shape, X_data_te.shape, y_tr.shape, y_te.shape
```

```python
y_tr = y_tr.as_matrix()

y_tr = y_tr.reshape((6,1))
y_tr = pd.DataFrame(y_tr)

type(y_tr), y_tr.shape
```

```python
xgb_model = xgb.XGBClassifier(n_estimators=100, max_depth=8, learning_rate=0.1, subsample=0.5)

xgb_model.fit(X='X_data_tr',y='y_tr')

xgb_model.score(X_data_tr,y_tr)
```

```python
GDS12_wNaN_new = GDS12_wNaN.iloc[:,0:11]
GDS12_wNaN_new['Label'] = Label
GDS12_wNaN_new = shuffle(GDS12_wNaN_new)
GDS12_wNaN_new 

GDS12_wNaN_new.to_csv("GDS12wNan_new.csv",index=False)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
X_data, y = GDS12_wNaN.drop('Label', axis=1), GDS12_wNaN.Label
seed = 7
test_size = 0.3
X_trian, X_test, y_train, y_test = train_test_split(X_data, y, test_size=test_size, random_state=seed)
X_data.shape, y.shape, X_trian.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
model1 = xgb.XGBClassifier()
model2 = xgb.XGBClassifier(n_estimators=100, max_depth=8, learning_rate=0.1, subsample=0.5)

train_model1 = model1.fit('X_trian', 'y_train')
train_model2 = model2.fit('X_trian', 'y_train')


pred1 = train_model1.predict(X_test)
pred2 = train_model2.predict(X_test)

print('Model 1 XGboost Report %r' % (classification_report(y_test, pred1)))
print('Model 2 XGboost Report %r' % (classification_report(y_test, pred2)))
```
